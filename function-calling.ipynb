````python
import requests
import json
import sys
import os
import re
````

````python
def get_response(user_input):
    url = "http://localhost:11434/api/generate"
    message = {
        "model": "llama3",
        "prompt": user_input,
    }
    response = requests.post(url, json=message)
    full_response_text = ""
    code_blocks = []

    if response.status_code == 200:
        # Aggregate response text
        response_lines = response.text.splitlines()
        for line in response_lines:
            try:
                line_json = json.loads(line)
                full_response_text += line_json["response"]
            except json.JSONDecodeError:
                print("JSON decode error in line:", line)
            
        print("Full response text:\n", full_response_text, "\n---\n")

        # Regex to find all code blocks marked by ```
        pattern = r"```(\w+)\s+(.*?)```"
        matches = re.findall(pattern, full_response_text, re.DOTALL)
        for match in matches:
            language, code = match
            code_blocks.append((language, code.strip()))

        # Print all extracted code blocks
        for language, code in code_blocks:
            print(f"Extracted {language} code:\n{code}\n---\n")

    else:
        print("Error:", response.status_code)
    return code_blocks
````

````python
code_blocks = get_response("So I have a task on data collecting from github, please use python to get the data from github api. Plesae highlight the code language when you give the code block.")
print("====")
for language, code in code_blocks:
    # Here you can further process or execute the extracted code if needed
    print(f"{language} code found:")
    print(code)
````
